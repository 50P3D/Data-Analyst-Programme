```python
import os
import time
import random
import pandas as pd
import fastf1 as ff1

# ------------------------------
# Cache configuration (CRITICAL)
# ------------------------------
CACHE_DIR = "../../data_raw/f1_cache"
CACHE_DIR = os.path.abspath(CACHE_DIR)

ff1.Cache.enable_cache(CACHE_DIR)
print(f"[FastF1] Cache directory set to: {CACHE_DIR}")

# ------------------------------
# What you want to run
# ------------------------------
YEARS = [2019, 2020, 2022]
SESSIONS = ["FP1", "FP2", "FP3"]  # use ["Q"] for lighter warmup

# ------------------------------
# Rate / pacing knobs
# ------------------------------
SLEEP_BETWEEN_SESSIONS = 20   # increase to 30–40 if throttled
JITTER = 3
MAX_RETRIES = 2
BACKOFF_BASE = 30             # retry backoff: 30s → 60s

PROGRESS_CSV = "fastf1_ingest_progress.csv"

# ------------------------------
# Helpers
# ------------------------------
def sleep_with_jitter(base: float, jitter: float) -> None:
    time.sleep(max(0.0, base + random.uniform(-jitter, jitter)))

def load_once(year: int, round_no: int, session_code: str):
    ses = ff1.get_session(year, round_no, session_code)
    ses.load(
        laps=True,
        weather=True,
        messages=False,
        telemetry=False
    )
    laps_n = int(len(ses.laps)) if (ses.laps is not None and not ses.laps.empty) else 0
    drivers_n = int(ses.laps["DriverNumber"].nunique()) if laps_n > 0 else 0
    return laps_n, drivers_n

# ------------------------------
# Load / initialize progress
# ------------------------------
if os.path.exists(PROGRESS_CSV):
    prog = pd.read_csv(PROGRESS_CSV)
else:
    prog = pd.DataFrame(
        columns=["year", "round", "session", "status", "laps", "drivers", "error"]
    )

done_keys = set(
    tuple(x)
    for x in prog.loc[prog["status"] == "OK", ["year", "round", "session"]].values
)

# ------------------------------
# Main loop
# ------------------------------
for year in YEARS:
    print(f"\n=== Year {year} ===")
    schedule = ff1.get_event_schedule(year)
    rounds = schedule["RoundNumber"].dropna().astype(int).tolist()

    for round_no in rounds:
        for session_code in SESSIONS:
            key = (year, round_no, session_code)
            if key in done_keys:
                continue

            print(f"Loading {year} R{round_no} {session_code}...")

            last_err = None
            ok = False
            laps_n = drivers_n = 0

            for attempt in range(1, MAX_RETRIES + 1):
                try:
                    laps_n, drivers_n = load_once(year, round_no, session_code)
                    ok = True
                    break
                except Exception as e:
                    last_err = e
                    print(
                        f"  FAIL attempt {attempt}/{MAX_RETRIES} "
                        f"-> {type(e).__name__}: {e}"
                    )
                    if attempt < MAX_RETRIES:
                        sleep_with_jitter(BACKOFF_BASE * attempt, jitter=5)

            # Record progress
            row = {
                "year": year,
                "round": round_no,
                "session": session_code,
                "status": "OK" if ok else "FAIL",
                "laps": laps_n,
                "drivers": drivers_n,
                "error": "" if ok else f"{type(last_err).__name__}: {last_err}",
            }
            prog = pd.concat([prog, pd.DataFrame([row])], ignore_index=True)
            prog.to_csv(PROGRESS_CSV, index=False)

            if ok:
                done_keys.add(key)
                print(f"  OK laps={laps_n} drivers={drivers_n}")
            else:
                print(f"  SKIPPED {year} R{round_no} {session_code}")

            sleep_with_jitter(SLEEP_BETWEEN_SESSIONS, JITTER)
```

```python
# def build_preQ_fp_features_for_weekend(year: int, rnd: int) -> pd.DataFrame:
#     fp_frames = []
#     meta = {"event_name": None, "country": None, "location": None}

#     # collect laps from all available FP sessions
#     for s in FP_SESSIONS:
#         try:
#             ses = fastf1.get_session(year, rnd, s)
#             ses.load(laps=True, weather=False, telemetry=False, messages=False)

#             # metadata from any available session
#             if meta["event_name"] is None:
#                 meta["event_name"] = getattr(ses.event, "EventName", None)
#                 meta["country"]    = getattr(ses.event, "Country", None)
#                 meta["location"]   = getattr(ses.event, "Location", None)

#             if ses.laps is None or ses.laps.empty:
#                 continue

#             df = clean_laps(ses.laps)

#             # standard join key = Driver (3-letter)
#             if "Driver" not in df.columns:
#                 continue

#             df["session"] = s
#             fp_frames.append(df)

#         except Exception:
#             # session might not exist for that weekend format
#             continue

#     if not fp_frames:
#         return pd.DataFrame(columns=ID_COLS + [
#             "n_fp_laps",
#             "n_fp_sessions_participated",
#             "best_fp_lap_time_overall",
#             "avg_fp_lap_time",
#             "median_fp_lap_time",
#             "best_last_fp_lap_time",
#             "best_last_fp_s1",
#             "best_last_fp_s2",
#             "best_last_fp_s3",
#         ])

#     fp = pd.concat(fp_frames, ignore_index=True)

#     # lap time seconds
#     fp["LapTime_s"] = fp["LapTime"].dt.total_seconds()

#     # last available FP session (FP3 else FP2 else FP1)
#     session_order = pd.Categorical(fp["session"], categories=["FP1","FP2","FP3"], ordered=True)
#     fp["session_ord"] = session_order
#     last_session = fp.groupby("Driver")["session_ord"].max()

#     # compute last-session bests by filtering per driver to their last session
#     fp_last = fp.merge(last_session.rename("last_s"), left_on="Driver", right_index=True)
#     fp_last = fp_last[fp_last["session_ord"] == fp_last["last_s"]].copy()

#     g = fp.groupby("Driver", dropna=False)

#     out = pd.DataFrame({
#         "driver": g.size().index.astype(str),
#         "n_fp_laps": g.size().values,
#         "n_fp_sessions_participated": g["session"].nunique().values,

#         # replacements for leaked Q-lap features:
#         "best_fp_lap_time_overall": g["LapTime_s"].min().values,
#         "avg_fp_lap_time": g["LapTime_s"].mean().values,
#         "median_fp_lap_time": g["LapTime_s"].median().values,
#     })

#     # last-session features
#     gl = fp_last.groupby("Driver", dropna=False)
#     out = out.merge(
#         pd.DataFrame({
#             "driver": gl.size().index.astype(str),
#             "best_last_fp_lap_time": gl["LapTime_s"].min().values,
#             "best_last_fp_s1": gl["Sector1Time"].min().dt.total_seconds().values if "Sector1Time" in fp_last.columns else np.nan,
#             "best_last_fp_s2": gl["Sector2Time"].min().dt.total_seconds().values if "Sector2Time" in fp_last.columns else np.nan,
#             "best_last_fp_s3": gl["Sector3Time"].min().dt.total_seconds().values if "Sector3Time" in fp_last.columns else np.nan,
#         }),
#         on="driver",
#         how="left"
#     )

#     out["year"] = year
#     out["round"] = rnd
#     out["event_name"] = meta["event_name"]
#     out["country"] = meta["country"]
#     out["location"] = meta["location"]

#     # order cols
#     out = out[["year","round","event_name","country","location","driver"] + [c for c in out.columns if c not in ["year","round","event_name","country","location","driver"]]]
#     return out

def build_preQ_fp_features_for_weekend(year: int, rnd: int) -> pd.DataFrame:
    fp_frames = []
    meta = {"event_name": None, "country": None, "location": None}

    # collect laps from all available FP sessions
    for s in FP_SESSIONS:
        try:
            ses = fastf1.get_session(year, rnd, s)
            ses.load(laps=True, weather=False, telemetry=False, messages=False)

            if meta["event_name"] is None:
                meta["event_name"] = getattr(ses.event, "EventName", None)
                meta["country"]    = getattr(ses.event, "Country", None)
                meta["location"]   = getattr(ses.event, "Location", None)

            if ses.laps is None or ses.laps.empty:
                continue

            df = clean_laps(ses.laps)
            if "Driver" not in df.columns:
                continue

            df = df.copy()
            df["session"] = s
            fp_frames.append(df)

        except Exception:
            continue

    # empty guard
    if not fp_frames:
        return pd.DataFrame(columns=ID_COLS + [
            "n_fp_laps",
            "n_fp_sessions_participated",
            "best_fp_lap_time_overall",
            "avg_fp_lap_time",
            "median_fp_lap_time",
            "best_last_fp_lap_time",
            "best_last_fp_s1",
            "best_last_fp_s2",
            "best_last_fp_s3",
        ])

    fp = pd.concat(fp_frames, ignore_index=True).reset_index(drop=True)

    # normalize key immediately (prevents Driver index/col ambiguity)
    fp = fp.rename(columns={"Driver": "driver"})
    fp["driver"] = fp["driver"].astype(str)

    # lap time seconds
    fp["LapTime_s"] = fp["LapTime"].dt.total_seconds()

    # last available FP session (FP3 else FP2 else FP1)
    fp["session_ord"] = pd.Categorical(fp["session"], categories=["FP1", "FP2", "FP3"], ordered=True)

    # last session per driver as a *column* (not index merge)
    last_s = (
        fp.groupby("driver", as_index=False)["session_ord"]
          .max()
          .rename(columns={"session_ord": "last_s"})
    )

    fp_last = fp.merge(last_s, on="driver", how="left")
    fp_last = fp_last[fp_last["session_ord"] == fp_last["last_s"]].copy()

    # base aggregations (all FP sessions)
    out = (
        fp.groupby("driver", as_index=False)
          .agg(
              n_fp_laps=("LapTime_s", "size"),
              n_fp_sessions_participated=("session", "nunique"),
              best_fp_lap_time_overall=("LapTime_s", "min"),
              avg_fp_lap_time=("LapTime_s", "mean"),
              median_fp_lap_time=("LapTime_s", "median"),
          )
    )

    # last-session aggregations
    last_agg = fp_last.groupby("driver", as_index=False).agg(
        best_last_fp_lap_time=("LapTime_s", "min"),
    )

    # sectors: compute only if present
    def _min_sector_seconds(df, col):
        if col not in df.columns:
            return None
        # sector cols are timedeltas; min() returns timedelta; convert to seconds
        s = df.groupby("driver")[col].min()
        return s.dt.total_seconds().rename(f"best_last_fp_{col.replace('Sector','s').replace('Time','').lower()}")

    # Build sector frames safely
    sec_frames = []
    if "Sector1Time" in fp_last.columns:
        s1 = fp_last.groupby("driver", as_index=False)["Sector1Time"].min()
        s1["best_last_fp_s1"] = s1["Sector1Time"].dt.total_seconds()
        sec_frames.append(s1[["driver", "best_last_fp_s1"]])
    if "Sector2Time" in fp_last.columns:
        s2 = fp_last.groupby("driver", as_index=False)["Sector2Time"].min()
        s2["best_last_fp_s2"] = s2["Sector2Time"].dt.total_seconds()
        sec_frames.append(s2[["driver", "best_last_fp_s2"]])
    if "Sector3Time" in fp_last.columns:
        s3 = fp_last.groupby("driver", as_index=False)["Sector3Time"].min()
        s3["best_last_fp_s3"] = s3["Sector3Time"].dt.total_seconds()
        sec_frames.append(s3[["driver", "best_last_fp_s3"]])

    # Merge last-session pieces
    out = out.merge(last_agg, on="driver", how="left")

    if sec_frames:
        sec = sec_frames[0]
        for nxt in sec_frames[1:]:
            sec = sec.merge(nxt, on="driver", how="outer")
        out = out.merge(sec, on="driver", how="left")
    else:
        # keep expected cols present
        out["best_last_fp_s1"] = np.nan
        out["best_last_fp_s2"] = np.nan
        out["best_last_fp_s3"] = np.nan

    # add meta
    out["year"] = year
    out["round"] = rnd
    out["event_name"] = meta["event_name"]
    out["country"] = meta["country"]
    out["location"] = meta["location"]

    # order cols
    key_cols = ["year", "round", "event_name", "country", "location", "driver"]
    out = out[key_cols + [c for c in out.columns if c not in key_cols]]

    # sanity: ensure no Driver ambiguity can ever occur downstream
    assert "Driver" not in out.columns
    assert "Driver" not in out.index.names

    return out
```

```python
def build_quali_labels_for_weekend(year: int, rnd: int) -> pd.DataFrame:
    ses = fastf1.get_session(year, rnd, Q_SESSION)
    ses.load(laps=False, weather=False, telemetry=False, messages=False)

    event_name = getattr(ses.event, "EventName", None)
    country    = getattr(ses.event, "Country", None)
    location   = getattr(ses.event, "Location", None)

    res = ses.results
    if res is None or res.empty:
        return pd.DataFrame(columns=["year","round","event_name","country","location","driver",TARGET])

    df = res.copy()
    # join key = abbreviation
    df = df.rename(columns={"Abbreviation": "driver", "Position": TARGET})

    out = df[["driver", TARGET]].copy()
    out["year"] = year
    out["round"] = rnd
    out["event_name"] = event_name
    out["country"] = country
    out["location"] = location

    return out[["year","round","event_name","country","location","driver",TARGET]]
```

```python
FEAT_PATH = OUT_DIR / "preq_features.parquet"
LAB_PATH  = OUT_DIR / "quali_labels.parquet"
DATA_PATH = OUT_DIR / "preq_dataset.parquet"

feat_rows = []
lab_rows = []
fail = []

for year in YEARS:
    try:
        sched = fastf1.get_event_schedule(year)
        rounds = sched["RoundNumber"].dropna().astype(int).tolist()
    except Exception as e:
        fail.append((year, None, "schedule_fail", repr(e)))
        continue

    for rnd in rounds:
        try:
            df_feat = build_preQ_fp_features_for_weekend(year, rnd)
            df_lab  = build_quali_labels_for_weekend(year, rnd)

            if not df_feat.empty:
                feat_rows.append(df_feat)
            if not df_lab.empty:
                lab_rows.append(df_lab)

            print(f"[OK] {year} R{rnd} | feat={len(df_feat)} | lab={len(df_lab)}")

        except Exception as e:
            fail.append((year, rnd, "weekend_fail", repr(e)))
            print(f"[FAIL] {year} R{rnd} -> {type(e).__name__}: {e}")

df_feat_all = pd.concat(feat_rows, ignore_index=True) if feat_rows else pd.DataFrame()
df_lab_all  = pd.concat(lab_rows, ignore_index=True) if lab_rows else pd.DataFrame()

print("feat:", df_feat_all.shape, "lab:", df_lab_all.shape, "fails:", len(fail))

df_feat_all.to_parquet(FEAT_PATH, index=False)
df_lab_all.to_parquet(LAB_PATH, index=False)

# merge dataset
df = df_feat_all.merge(df_lab_all, on=["year","round","event_name","country","location","driver"], how="inner")
df.to_parquet(DATA_PATH, index=False)

print("Saved:")
print(" -", FEAT_PATH)
print(" -", LAB_PATH)
print(" -", DATA_PATH)

df.head()
```
